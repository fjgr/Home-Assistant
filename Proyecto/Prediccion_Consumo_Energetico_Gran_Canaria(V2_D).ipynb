## 5. Optimización, normalización y calidad del set de datos

Antes de proceder con el modelado, es necesario preparar adecuadamente los datos, lo que incluye la creación de características temporales, la normalización de variables y la división en conjuntos de entrenamiento y prueba.

```python
# Crear características temporales
df['año'] = df['fecha'].dt.year
df['mes'] = df['fecha'].dt.month
df['dia_semana'] = df['fecha'].dt.dayofweek
df['dia_año'] = df['fecha'].dt.dayofyear
df['trimestre'] = df['fecha'].dt.quarter

# Crear variables estacionales (seno y coseno para capturar ciclicidad)
df['sin_mes'] = np.sin(2 * np.pi * df['mes']/12)
df['cos_mes'] = np.cos(2 * np.pi * df['mes']/12)
df['sin_dia_año'] = np.sin(2 * np.pi * df['dia_año']/365)
df['cos_dia_año'] = np.cos(2 * np.pi * df['dia_año']/365)

# Preparar datos para modelado por zonas
zonas = ['norte', 'este', 'sur', 'resto']
X_zonas = {}
y_zonas = {}
X_train_zonas = {}
X_test_zonas = {}
y_train_zonas = {}
y_test_zonas = {}
scalers = {}

# Definir características para cada zona
for zona in zonas:
    # Seleccionar características relevantes para cada zona
    if zona == 'norte':
        features = ['temp_norte', 'hum_norte', 'año', 'mes', 'dia_semana', 'trimestre',
                   'sin_mes', 'cos_mes', 'sin_dia_año', 'cos_dia_año']
        target = 'consumo_norte'
    elif zona == 'este':
        features = ['temp_este', 'hum_este', 'año', 'mes', 'dia_semana', 'trimestre',
                   'sin_mes', 'cos_mes', 'sin_dia_año', 'cos_dia_año']
        target = 'consumo_este'
    elif zona == 'sur':
        features = ['temp_sur', 'hum_sur', 'año', 'mes', 'dia_semana', 'trimestre',
                   'sin_mes', 'cos_mes', 'sin_dia_año', 'cos_dia_año']
        target = 'consumo_sur'
    else:  # resto
        features = ['temp_resto', 'hum_resto', 'año', 'mes', 'dia_semana', 'trimestre',
                   'sin_mes', 'cos_mes', 'sin_dia_año', 'cos_dia_año']
        target = 'consumo_resto'

    # Preparar X e y
    X_zonas[zona] = df[features]
    y_zonas[zona] = df[target]

    # Dividir en conjuntos de entrenamiento y prueba (80% train, 20% test)
    X_train_zonas[zona], X_test_zonas[zona], y_train_zonas[zona], y_test_zonas[zona] = train_test_split(
        X_zonas[zona], y_zonas[zona], test_size=0.2, random_state=42, shuffle=False
    )

    # Normalizar características
    scalers[zona] = StandardScaler()
    X_train_zonas[zona] = pd.DataFrame(
        scalers[zona].fit_transform(X_train_zonas[zona]),
        columns=features,
        index=X_train_zonas[zona].index
    )
    X_test_zonas[zona] = pd.DataFrame(
        scalers[zona].transform(X_test_zonas[zona]),
        columns=features,
        index=X_test_zonas[zona].index
    )

# Visualizar división de datos de entrenamiento y prueba
plt.figure(figsize=(14, 10))
for i, zona in enumerate(zonas, 1):
    plt.subplot(2, 2, i)
    plt.plot(df['fecha'].iloc[y_train_zonas[zona].index], y_train_zonas[zona],
             label='Entrenamiento', alpha=0.7)
    plt.plot(df['fecha'].iloc[y_test_zonas[zona].index], y_test_zonas[zona],
             label='Prueba', alpha=0.7)
    plt.title(f'División de datos para zona {zona.capitalize()}')
    plt.xlabel('Fecha')
    plt.ylabel(f'Consumo {zona} (MWh)')
    plt.legend()
    plt.grid(True)
plt.tight_layout()
plt.show()

# Información sobre los conjuntos de datos
print('Información de los conjuntos de datos:')
for zona in zonas:
    print(f'\nZona {zona.capitalize()}:')
    print(f'  Características: {X_zonas[zona].columns.tolist()}')
    print(f'  Muestras de entrenamiento: {X_train_zonas[zona].shape[0]}')
    print(f'  Muestras de prueba: {X_test_zonas[zona].shape[0]}')
    print(f'  Rango de fechas entrenamiento: {df.iloc[X_train_zonas[zona].index].fecha.min().strftime("%d/%m/%Y")} - {df.iloc[X_train_zonas[zona].index].fecha.max().strftime("%d/%m/%Y")}')
    print(f'  Rango de fechas prueba: {df.iloc[X_test_zonas[zona].index].fecha.min().strftime("%d/%m/%Y")} - {df.iloc[X_test_zonas[zona].index].fecha.max().strftime("%d/%m/%Y")}')
```

El proceso de preparación de datos incluye varias etapas importantes:

1. **Creación de características temporales**: Se han generado variables que capturan la información temporal (año, mes, día de la semana, día del año, trimestre), lo que permite a los modelos identificar patrones estacionales y tendencias.

2. **Variables cíclicas**: Para representar adecuadamente la naturaleza cíclica del tiempo, se han creado transformaciones sinusoidales del mes y del día del año, lo que evita el problema de discontinuidad en estas variables (por ejemplo, diciembre y enero son consecutivos pero numéricamente distantes).

3. **División por zonas**: Se han preparado conjuntos de datos específicos para cada zona geográfica, seleccionando las variables meteorológicas correspondientes a cada región.

4. **División temporal**: Los datos se han dividido en conjuntos de entrenamiento (80%) y prueba (20%) de forma cronológica, sin aleatorización, para mantener la integridad temporal de las series y evaluar la capacidad de los modelos para predecir períodos futuros.

5. **Normalización**: Las variables predictoras se han normalizado utilizando StandardScaler, lo que mejora el rendimiento de muchos algoritmos de aprendizaje automático al equiparar la escala de las diferentes características.

La visualización de la división de datos muestra claramente la separación cronológica entre los conjuntos de entrenamiento y prueba, con el conjunto de prueba abarcando aproximadamente los últimos 10 meses del período analizado (agosto 2023 - junio 2024).

## 6. Desarrollo del modelo predictivo

### 6.1. Justificación del modelo seleccionado

Para abordar el problema de predicción del consumo energético en Gran Canaria, se ha optado por un enfoque multirramificado que utiliza tres tipos de modelos diferentes:

1. **Regresión Lineal**: Se selecciona como modelo base por su simplicidad, interpretabilidad y eficiencia computacional. Aunque puede no capturar relaciones no lineales complejas, proporciona una línea base útil para comparar con modelos más sofisticados.

2. **Random Forest**: Este modelo de conjunto basado en árboles de decisión se elige por su capacidad para capturar relaciones no lineales, manejar diferentes tipos de variables y proporcionar información sobre la importancia de las características. Además, es menos propenso al sobreajuste que los árboles individuales.

3. **Redes Recurrentes (LSTM)**: Se seleccionan por su capacidad para modelar dependencias temporales y patrones secuenciales en los datos, lo que es relevante para series temporales como el consumo energético. Las LSTM (Long Short-Term Memory) son particularmente adecuadas para capturar dependencias a largo plazo.

El enfoque multirramificado por subzonas se justifica por las diferencias geográficas y climáticas entre las distintas regiones de Gran Canaria, así como por los diferentes patrones de consumo asociados a cada zona (turística, residencial, etc.). Este enfoque permite desarrollar modelos específicos que capturan mejor las particularidades de cada región, mejorando la precisión global de la predicción.

### 6.2. Descripción del modelo

A continuación, se implementan los tres tipos de modelos para cada una de las zonas geográficas definidas.

```python
# Función para entrenar y evaluar modelos
def entrenar_evaluar_modelo(modelo, nombre_modelo, zona):
    # Entrenar modelo
    modelo.fit(X_train_zonas[zona], y_train_zonas[zona])

    # Predecir
    y_pred_train = modelo.predict(X_train_zonas[zona])
    y_pred_test = modelo.predict(X_test_zonas[zona])

    # Calcular métricas
    mae_train = mean_absolute_error(y_train_zonas[zona], y_pred_train)
    rmse_train = np.sqrt(mean_squared_error(y_train_zonas[zona], y_pred_train))
    r2_train = r2_score(y_train_zonas[zona], y_pred_train)

    mae_test = mean_absolute_error(y_test_zonas[zona], y_pred_test)
    rmse_test = np.sqrt(mean_squared_error(y_test_zonas[zona], y_pred_test))
    r2_test = r2_score(y_test_zonas[zona], y_pred_test)

    # Guardar resultados
    resultados = {
        'modelo': nombre_modelo,
        'zona': zona,
        'mae_train': mae_train,
        'rmse_train': rmse_train,
        'r2_train': r2_train,
        'mae_test': mae_test,
        'rmse_test': rmse_test,
        'r2_test': r2_test,
        'y_pred_test': y_pred_test
    }

    return resultados, y_pred_test

# 1. Modelos de Regresión Lineal
print('Entrenando modelos de Regresión Lineal...')
resultados_lr = []
predicciones_lr = {}

for zona in zonas:
    print(f'\nZona {zona.capitalize()}:')
    modelo_lr = LinearRegression()
    resultado, pred = entrenar_evaluar_modelo(modelo_lr, 'lr', zona)
    resultados_lr.append(resultado)
    predicciones_lr[zona] = pred
    print(f'  MAE test: {resultado["mae_test"]:.2f}')
    print(f'  RMSE test: {resultado["rmse_test"]:.2f}')
    print(f'  R² test: {resultado["r2_test"]:.4f}')

# 2. Modelos Random Forest
print('\nEntrenando modelos Random Forest...')
resultados_rf = []
predicciones_rf = {}

for zona in zonas:
    print(f'\nZona {zona.capitalize()}:')
    modelo_rf = RandomForestRegressor(n_estimators=100, random_state=42)
    resultado, pred = entrenar_evaluar_modelo(modelo_rf, 'rf', zona)
    resultados_rf.append(resultado)
    predicciones_rf[zona] = pred
    print(f'  MAE test: {resultado["mae_test"]:.2f}')
    print(f'  RMSE test: {resultado["rmse_test"]:.2f}')
    print(f'  R² test: {resultado["r2_test"]:.4f}')
```

Ahora implementaremos el modelo de red recurrente (LSTM) para cada zona:

```python
# Definir la arquitectura de la red recurrente (LSTM)
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        # LSTM layer
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)

        # Fully connected layer
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        # Initialize hidden state with zeros
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        # Forward propagate LSTM
        out, _ = self.lstm(x, (h0, c0))

        # Decode the hidden state of the last time step
        out = self.fc(out[:, -1, :])
        return out

# Función para preparar datos para LSTM (secuencias)
def prepare_sequences(X, y, seq_length=7):
    X_seq, y_seq = [], []
    for i in range(len(X) - seq_length):
        X_seq.append(X[i:i+seq_length])
        y_seq.append(y[i+seq_length])
    return np.array(X_seq), np.array(y_seq)

# Función para entrenar y evaluar modelo LSTM
def entrenar_evaluar_lstm(zona, seq_length=7, hidden_size=64, num_layers=2,
                         batch_size=32, learning_rate=0.001, epochs=50):
    # Cargar datos
    X_train = X_train_zonas[zona].values
    X_test = X_test_zonas[zona].values
    y_train = y_train_zonas[zona].values
    y_test = y_test_zonas[zona].values

    # Preparar secuencias
    X_train_seq, y_train_seq = prepare_sequences(X_train, y_train, seq_length)
    X_test_seq, y_test_seq = prepare_sequences(X_test, y_test, seq_length)

    # Convertir a tensores
    X_train_tensor = torch.FloatTensor(X_train_seq)
    y_train_tensor = torch.FloatTensor(y_train_seq).view(-1, 1)
    X_test_tensor = torch.FloatTensor(X_test_seq)
    y_test_tensor = torch.FloatTensor(y_test_seq).view(-1, 1)

    # Crear datasets y dataloaders
    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

    # Definir modelo, criterio y optimizador
    input_size = X_train.shape[1]  # Número de características
    output_size = 1  # Predicción de consumo

    model = LSTMModel(input_size, hidden_size, num_layers, output_size)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    # Entrenamiento
    train_losses = []
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        for inputs, targets in train_loader:
            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, targets)

            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        epoch_loss = running_loss / len(train_loader)
        train_losses.append(epoch_loss)

        if (epoch+1) % 10 == 0:
            print(f'Zona {zona}, Epoch [{epoch+1}/{epochs}], Loss: {epoch_loss:.4f}')

    # Evaluación
    model.eval()
    with torch.no_grad():
        y_pred_train = model(X_train_tensor).numpy().flatten()
        y_pred_test = model(X_test_tensor).numpy().flatten()

    # Calcular métricas
    mae_train = mean_absolute_error(y_train_seq, y_pred_train)
    rmse_train = np.sqrt(mean_squared_error(y_train_seq, y_pred_train))
    r2_train = r2_score(y_train_seq, y_pred_train)

    mae_test = mean_absolute_error(y_test_seq, y_pred_test)
    rmse_test = np.sqrt(mean_squared_error(y_test_seq, y_pred_test))
    r2_test = r2_score(y_test_seq, y_pred_test)

    # Guardar resultados
    resultados = {
        'modelo': 'lstm',
        'zona': zona,
        'mae_train': mae_train,
        'rmse_train': rmse_train,
        'r2_train': r2_train,
        'mae_test': mae_test,
        'rmse_test': rmse_test,
        'r2_test': r2_test
    }

    # Visualizar pérdida durante entrenamiento
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses)
    plt.title(f'Pérdida durante entrenamiento - LSTM Zona {zona.capitalize()}')
    plt.xlabel('Época')
    plt.ylabel('Pérdida (MSE)')
    plt.grid(True)
    plt.show()

    return resultados, y_pred_test, y_test_seq

# 3. Modelos LSTM
print('\nEntrenando modelos LSTM...')
resultados_lstm = []
predicciones_lstm = {}
y_test_seq_zonas = {}

for zona in zonas:
    print(f'\nZona {zona.capitalize()}:')
    resultado, pred, y_test_seq = entrenar_evaluar_lstm(zona, epochs=50)
    resultados_lstm.append(resultado)
    predicciones_lstm[zona] = pred
    y_test_seq_zonas[zona] = y_test_seq
    print(f'  MAE test: {resultado["mae_test"]:.2f}')
    print(f'  RMSE test: {resultado["rmse_test"]:.2f}')
    print(f'  R² test: {resultado["r2_test"]:.4f}')
```

```python
# Crear visualización comparativa adicional
plt.figure(figsize=(15, 10))
for i, zona in enumerate(zonas):
    plt.subplot(2, 2, i+1)
    df_zona = df_resultados[df_resultados['zona'] == zona]

    # Ordenar por RMSE (menor es mejor)
    df_zona = df_zona.sort_values('rmse_test')

    # Crear barras para RMSE
    plt.bar(df_zona['modelo'], df_zona['rmse_test'], color=['blue', 'green', 'red'])
    plt.title(f'Comparación de RMSE - Zona {zona.capitalize()}')
    plt.ylabel('RMSE')
    plt.grid(True, alpha=0.3)

    # Añadir valores en las barras
    for j, v in enumerate(df_zona['rmse_test']):
        plt.text(j, v + 5, f'{v:.1f}', ha='center')

plt.tight_layout()
plt.show()

# Crear gráfico de radar para comparar modelos
def radar_chart(df, metrics, title):
    # Número de variables
    N = len(metrics)

    # Ángulos para cada eje
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]  # Cerrar el polígono

    # Crear figura
    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

    # Para cada modelo
    for i, modelo in enumerate(df['modelo'].unique()):
        values = df[df['modelo'] == modelo][metrics].values.flatten().tolist()
        values += values[:1]  # Cerrar el polígono

        # Dibujar el polígono
        ax.plot(angles, values, linewidth=2, linestyle='solid', label=modelo)
        ax.fill(angles, values, alpha=0.1)

    # Etiquetas
    plt.xticks(angles[:-1], metrics, size=12)

    # Título y leyenda
    plt.title(title, size=15, y=1.1)
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

    plt.tight_layout()
    plt.show()

# Preparar datos para el gráfico de radar
# Normalizar métricas para que estén en la misma escala (0-1)
radar_df = df_resultados.copy()

# Para MAE y RMSE, menor es mejor, así que invertimos
for metric in ['mae_test', 'rmse_test']:
    max_val = radar_df[metric].max()
    radar_df[f'{metric}_norm'] = 1 - (radar_df[metric] / max_val)

# Para R², mayor es mejor
min_r2 = radar_df['r2_test'].min()
max_r2 = radar_df['r2_test'].max()
if min_r2 < 0:
    # Ajustar para valores negativos
    radar_df['r2_test_norm'] = (radar_df['r2_test'] - min_r2) / (max_r2 - min_r2)
else:
    radar_df['r2_test_norm'] = radar_df['r2_test'] / max_r2

# Crear gráfico de radar para cada zona
for zona in zonas:
    df_zona = radar_df[radar_df['zona'] == zona]
    radar_chart(
        df_zona,
        ['mae_test_norm', 'rmse_test_norm', 'r2_test_norm'],
        f'Comparación de Modelos - Zona {zona.capitalize()}'
    )

# Predicción del consumo total
print('\nPredicción del consumo total de la isla:')
print('Para predecir el consumo total, se suman las predicciones de los mejores modelos para cada zona.')

# Identificar el mejor modelo para cada zona
mejores_modelos = {}
for zona in zonas:
    df_zona = df_resultados[df_resultados['zona'] == zona]
    mejor_modelo = df_zona.loc[df_zona['rmse_test'].idxmin()]['modelo']
    mejores_modelos[zona] = mejor_modelo

# Obtener predicciones del mejor modelo para cada zona
predicciones_mejores = {}
for zona in zonas:
    if mejores_modelos[zona] == 'lr':
        predicciones_mejores[zona] = predicciones_lr[zona]
    elif mejores_modelos[zona] == 'rf':
        predicciones_mejores[zona] = predicciones_rf[zona]
    else:  # lstm
        # Ajustar para la diferencia en longitud debido a secuencias
        pred_lstm = np.zeros(len(predicciones_lr[zona]))
        pred_lstm[7:] = predicciones_lstm[zona]
        predicciones_mejores[zona] = pred_lstm

# Calcular predicción del consumo total
consumo_total_pred = np.zeros(len(predicciones_lr['norte']))
for zona in zonas:
    consumo_total_pred += predicciones_mejores[zona]

# Comparar con consumo total real
consumo_total_real = df['consumo_total'].iloc[y_test_zonas['norte'].index].values
fechas_test = df['fecha'].iloc[y_test_zonas['norte'].index].values

# Calcular métricas para el consumo total
mae_total = mean_absolute_error(consumo_total_real, consumo_total_pred)
rmse_total = np.sqrt(mean_squared_error(consumo_total_real, consumo_total_pred))
r2_total = r2_score(consumo_total_real, consumo_total_pred)

print(f'\nMétricas para la predicción del consumo total:')
print(f'MAE: {mae_total:.2f} MWh')
print(f'RMSE: {rmse_total:.2f} MWh')
print(f'R²: {r2_total:.4f}')

# Visualizar predicción del consumo total
plt.figure(figsize=(12, 6))
plt.plot(fechas_test, consumo_total_real, label='Consumo total real', color='black', alpha=0.7)
plt.plot(fechas_test, consumo_total_pred, label='Consumo total predicho', color='blue', alpha=0.7)
plt.title('Predicción del Consumo Energético Total de Gran Canaria')
plt.xlabel('Fecha')
plt.ylabel('Consumo Total (MWh)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
